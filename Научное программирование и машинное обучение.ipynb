```python
#5) среднее значение, медиана, мода (вычисляем только для дискретных переменных), 
#размах, дисперсия, среднеквадратичное отклонение, 1й/2й/3й квартили, межквартильный размах

def descriptive_statistic(x, discret):
    result = {}
    result['среднее'] = np.nanmean(x)
    result['медиана'] = np.nanmedian(x)
    if discret == 'yes':
        result['мода'] = st.mode(x)
    result['размах'] = np.nanmax(x) - np.nanmin(x)
    result['дисперсия'] = np.nanvar(x, ddof = 1)
    result['среднеквадратичное отклонение'] = np.nanstd(x, ddof = 1)
    result['квартили 1/2/3'] = np.nanquantile(x, [0.25, 0.5, 0.75])
    result['межквартильный размах'] = result['квартили 1/2/3'][2] - result['квартили 1/2/3'][0]
    
    return result


lst_x = {'age' : 'yes', 'trestbps' : 'yes', 'chol' : 'yes', 'thalach' : 'yes', 'oldpeak' : 'no'}
dict_with_statistic = {}

for x in lst_x.keys():
    dict_with_statistic[x] = descriptive_statistic(df[x], lst_x[x])

df_statistic = pd.DataFrame(dict_with_statistic)
df_statistic
```

Н0 среднее значение выборки а = среднему значению b 

1 - Н1 среднее значение выборок отличаются

2 - Н1 среднее значение в выборке а меньше, чем b

все объяснения указаны в функции -> и выводятся на экран в каждом случае

```python
def choose_method(a, b, alpha = 0.05, dependence = False):
    
    norm1 = scist.shapiro(a)[1]
    norm2 = scist.shapiro(b)[1]
    
    #если нормальное распределение
    if norm1 > alpha and norm2 > alpha:
        print(f'Распределения нормальные \np-value a: {norm1} b: {norm2}')
        
        # критерий Бартлета для сравнения дисперсий
        bart = scist.bartlett(a, b)[1]
        if bart > alpha:
            print(f'дисперсии примерно равны p-value - {bart}')
            
            #если независимые
            if not dependence:
                print('Выборки независимые')
                print('Метод t-критерий Стьюдента')
                print('two - sided: ', scist.ttest_ind(a, b)[1], 'less: ', scist.ttest_ind(a, b, alternative='less')[1], 'greater: ',scist.ttest_ind(a, b, alternative='greater')[1] )
           
            #зависимые
            else:
                print('Выборки зависимые')
                print('Метод t-критерий Стьюдента для зависимых выборок')
                print('two - sided: ', scist.ttest_rel(a, b)[1], 'less: ', scist.ttest_rel(a, b, alternative='less')[1], 'greater: ', scist.ttest_rel(a, b, alternative='greater')[1])


        #если дисперсии не равны
        else:
            print(f'дисперсии не равны p-value - {bart}')
            if not dependence:
                print('Выборки независимые')
                print('Метод t-критерий Уэлча')
                print('two - sided: ', scist.ttest_ind(a, b, equal_var=False)[1], 'less: ', scist.ttest_ind(a, b, equal_var=False, alternative='less')[1], 'greater: ',scist.ttest_ind(a, b, equal_var=False, alternative='greater')[1] )
            else:
                print('Выборки зависимые')
                print('Метод t-критерий Стьюдента для зависимых выборок с неравными дисперсиями')
                print('two - sided: ', scist.ttest_rel(a, b, equal_var=False)[1], 'less: ', scist.ttest_rel(a, b, equal_var=False, alternative='less')[1], 'greater: ',scist.ttest_rel(a, b, equal_var=False, alternative='greater')[1] )
    
    #если распределение не нормальное
    else:
        if not dependence:
            print('Выборки независимые')
            print(f'Распределения a или/и b не нормальные\n p-value a: {norm1} b: {norm2}')
            print('Метод Манна-Уитни')
            print('two - sided: ', scist.mannwhitneyu(a, b)[1], 'less: ', scist.mannwhitneyu(a, b, alternative='less')[1], 'greater: ',scist.mannwhitneyu(a, b, alternative='greater')[1]) 
        else:
            print('Выборки зависимые')
            print(f'Распределения a или/и b не нормальные\n p-value a: {norm1} b: {norm2}')
            print('Критерий Уилкоксона')
            print('two - sided: ', scist.wilcoxon(a, b)[1], 'less: ', scist.wilcoxon(a, b, alternative='less')[1], 'greater: ',scist.wilcoxon(a, b, alternative='greater')[1]) 
```

```python
def FDR_corrections(p, alpha = 0.05):
    sort_p = sorted(p)
    p_sorted_id = np.argsort(p)
    m = len(sort_p)
    list_fdr = []
    
    i = 0
    while i < m:
        if sort_p[i] >= (i+1)*alpha/m:
            for x in range(m - i):
                list_fdr.append(False)
                i = m
        else:
            i += 1
            list_fdr.append(True)
            
            
    res_fdr = [i for i in range(m)]
    res_q = [i for i in range(m)]
    q_value = [(sort_p[i] * m/ (i + 1)) for i in range(m)]
    
    y = 0
    for x in p_sorted_id:
        res_fdr[x] = list_fdr[y]
        res_q[x] = q_value[y]
        y+=1
        
    return res_fdr, res_q
```

Необходимо написать функцию, которая решает СЛАУ. 

Входные аргументы: матрица системы 𝐴 и вектор правых частей 𝑏.  

Внутри функции необходимо определить, имеет ли система решение (исследовать ранги - для вычисления рангов использовать функцию из numpy), а если имеет, то сколько, и в зависимости от этого использовать тот или иной метод поиска решения из библиотеки numpy. 

Функция в зависимости от входных данных должна возвращать либо None, если решений нет; одно точное решение или одно из множества решений.

Протестировать функцию для решения СЛАУ, представленных на рис. СЛАУ.jpg

```python
import numpy.linalg as linalg
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solver_SLAE(A, b):
    extended_matrix = A.copy()
    for i in range(len(A)):
        extended_matrix[i].append(b[i])
    rang = linalg.matrix_rank(A)
    if  rang == linalg.matrix_rank(extended_matrix):
        #совместная система
        if rang == len(A[0]):
            #единственное решение - система определена
            if len(A[0]) == len(A): #если квадратная система 
                return linalg.solve(A, b)
            else:
                return linalg.lstsq(A, b, rcond=None)[0]
        elif rang < len(A[0]):
            #множество решений - неопределена
            return linalg.lstsq(A, b, rcond=None)[0]
    else:
        #нет решений - несовместная
        return None
```

```python
matrix1 = [[4, -2], [6, 1]]
b1 = [22, 45]

matrix2 = [[7, -2, -1], [6, -4, -5], [1, 2, 4]]
b2 = [2, 3, 5]

matrix3 = [[2, 3, -1, 1], [8, 12, -9, 8], [4, 6, 3, -2], [2, 3, 9, -7]]
b3 = [1, 3, 3, 3]

x1 = solver_SLAE(matrix1, b1)
print('Solution = ', x1)
print(np.allclose(np.dot(matrix1, x1), b1, atol=1e-08))

x2 = solver_SLAE(matrix2, b2)
print('Solution = ', x2)
print(np.allclose(np.dot(matrix2, x2), b2, atol=1e-08))

x3 = solver_SLAE(matrix3, b3)
print('Solution = ', x3)
print(np.allclose(np.dot(matrix3, x3), b3, atol=1e-08))
```

```python
def metod_Euler(fun_sys, time_s_e, initial_values, h, args):
    times = np.arange(time_s_e[0], time_s_e[1]+h, h)
    n = len(times) 
    m = len(initial_values)
    y = np.zeros((n, m))
    y[0,:] = initial_values
    for i in range(1, n):
        y[i, :] = y[i-1, :] + h * fun_sys(times[i-1], y[i-1, :], *args)
    return times, y
```

```python
def metod_Runge_Kutta(fun_sys, time_s_e, initial_values, h, args):
    times = np.arange(time_s_e[0], time_s_e[1]+h, h)
    n = len(times) 
    m = len(initial_values)
    y = np.zeros((n, m))
    y[0,:] = initial_values
    for i in range(1, n):
        k = [1,2,3,4]
        k[0] = fun_sys(times[i-1], y[i-1, :], *args)
        for j in range(1, 3):
            k[j] = fun_sys(np.add(times[i-1], h/2), np.add(y[i-1, :], np.dot(h/2, k[j-1])), *args)
        k[3] = fun_sys(times[i-1] + h, np.add(y[i-1, :], np.dot(h, k[2])), *args)
        y[i, :] = y[i-1, :] + h/6 * (k[0] + 2*k[1] + 2*k[2] + k[3])
    return times, y
```

```python
def method_Monte_Carlo(i_x, N = 100, n_realize = 10000):
    life_times = []
    for r in range(0, n_realize):
        i = i_x
        n_steps = 0
        while i != 0 and i != N:
            p = i / N
            ksi1, ksi2 = np.random.uniform(0, 1, 2)
            if ksi1 <= p and ksi2 > p:
                i += 1
            elif ksi1 > p and ksi2 <= p:
                i -= 1
            n_steps += 1
        life_times.append(n_steps)
    return life_times
```

<strong>P.S * В общем после этого задания стало понятно насколько удобно все считать не "руками" :) поэтому я постаралась написать простейший алгоритм байесовского классификатора. На тестовых выборках вроде работает

```python
#должна быть таблица, где есть столбик классы, затем столбцы с признаками и количеством
#lst_sign словарь признаков в виде {признак: 0, 1}
#сил хватило на реализацию только: наличие отсуствия признака - 0 и 1

def bayesia_classifier(table_frequency,lst_sign , s_j = 2):
    #вероятности
    frec = {}
    sum_count = table_frequency['Общее кол-во'].iloc[table_frequency.shape[0] - 1]
    for i in range(table_frequency.shape[0] - 1):
        frec[table_frequency['Класс'].iloc[i]] = table_frequency['Общее кол-во'].iloc[i] / sum_count    
    #print(f'Априорные вероятности классов - {frec}')
    
    #вероятности с поправкой Лапласа
    laplas = {}
    for sign in table_frequency.columns:
        if sign != 'Класс' and sign != 'Общее кол-во':
            temp = {}
            for var in range(table_frequency.shape[0] - 1):
                if s_j == 2:
                    temp[table_frequency['Класс'].iloc[var]] = {1:(int(table_frequency[sign].iloc[var]) + 1) / (int(table_frequency['Общее кол-во'].iloc[var]) + 2),
                                                   0: 1 - (int(table_frequency[sign].iloc[var]) + 1) / (int(table_frequency['Общее кол-во'].iloc[var]) + 2)}
            laplas[sign] = temp
    #print('\n условные вероятности с поправкой Лапласа')
    #print(laplas)
    
    pr = {}
    for x in range(table_frequency.shape[0] - 1):
        temp_2 = 1
        cl = table_frequency['Класс'].iloc[x]
        for s in lst_sign:
            temp_2 *= laplas[s][cl][lst_sign[s]]
        pr[cl] = temp_2 * frec[cl]
        
    #print(pr)
    return max(pr, key = pr.get)
```

```python
lst_s_1 = {'Длинный': 0,'Сладкий': 1,'Желтый': 1}
lst_s_2 = {'Длинный': 0,'Сладкий': 1,'Желтый': 0}
lst_s_3 = {'Длинный': 1,'Сладкий': 1,'Желтый': 0}

print(bayesia_classifier(df, lst_s_1))
print(bayesia_classifier(df, lst_s_2))
print(bayesia_classifier(df, lst_s_3))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets  
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, f1_score, RocCurveDisplay, roc_auc_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
X, y = datasets.load_breast_cancer(return_X_y=True) 
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state = 777)
```

```python
# В range и так не входит последний индекс, поэтому лишним было вычитать ещё единицу
# range(1, len(X_train) - 1) - от 1 до 396, а объектов в обучающей выборке - 398, поэтому должно быть от 1 до 397

# Любой классификатор обучается только на обучающей выборке! nbc.fit(X_test, y_test) - грубая ошибка

# Не подписаны оси у графика

#поправила!
```

```python
def task_1_2(X_train, X_test, y_train, y_test):
    fig, ax = plt.subplots(figsize = (18, 6))

    k = []
    errors_train = []
    errors_test = []
    for i in range(1, len(X_train)):

        k.append(i)
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(X_train, y_train)

        y_train_predict = knn.predict(X_train)
        y_test_predict = knn.predict(X_test)

        error_train = np.mean(y_train != y_train_predict)
        errors_train.append(error_train)
        error_test = np.mean(y_test != y_test_predict)
        errors_test.append(error_test)

    nbc = GaussianNB()
    nbc.fit(X_train, y_train)
    nbc_error = np.mean(y_test != nbc.predict(X_test)) 

    ax.plot(k, errors_train, label = 'train')
    ax.plot(k, errors_test, label = 'test')
    ax.plot(k, [nbc_error for k in range(len(k))], '--g' , label = 'GaussianNB')
    ax.scatter(errors_test.index(min(errors_test)) + 1, min(errors_test), color = 'red')
    ax.set_xlabel('k')
    ax.set_ylabel('ошибка')
    ax.legend()
    plt.show()
    return errors_test
```

```python
err1 = task_1_2(X_train, X_test, y_train, y_test)
```

<strong>Вывод:</strong> при маленьком k ошибка очень маленькая на тренировочной выборке, но больше на тестовой, так как у каждого образца минимум ближайших соседей. при k больше 250 слишком далекие образцы становятся соседями и классифакатор плохо распознает как тренировочную, так и тестовую выборку. А вот при k от 8 до 15 показывает оптимальную обучаемость, лучшим k является 8.

```python
k = err1.index(min(err1)) + 1
kNN = KNeighborsClassifier(n_neighbors=k)
kNN.fit(X_train, y_train)
y_test_predict = kNN.predict(X_test)

print(f'accuracy: {accuracy_score(y_test, y_test_predict)}')
print(f'f1: {f1_score(y_test, y_test_predict)}')
print(f'AUC: {roc_auc_score(y_test, kNN.predict_proba(X_test)[:,1])}')

fig, ax = plt.subplots()
RocCurveDisplay.from_estimator(kNN, X_test, y_test, ax=ax)
plt.title('ROC-кривая')
plt.show()
#P.S AUC также отображается на графике
```

```python
# -
def first_step_find_duplicates(df):
    df_copy = df.copy()
    categorical_columns = [c for c in df_copy.columns if df_copy[c].dtype.name == 'object']
    for c in categorical_columns:
        df_copy[c] = [str(prz).upper() for prz in df_copy[c]]
        df_copy = df_copy.replace({c: {'NAN':np.NaN} })
    df_copy.drop_duplicates(subset=None, keep='first', inplace=True,
                          ignore_index=False)
    return df_copy
```

```python
# +
def second_step_noninfor_sign(df):
    df_copy = df.copy()
    n = len(df_copy)
    for col in df_copy.columns:
        count = {}
        for val in df_copy[col]:
            if val in count:
                count[val] += 1
            else:
                count[val] = 1
        for k in count.keys():
            if count[k] > n * 0.95:
                df_copy.pop(col)
    return df_copy
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, RocCurveDisplay, roc_auc_score
from sklearn.decomposition import PCA
from sklearn.model_selection import LeaveOneOut
```

```python
from sklearn import datasets  
X, y = datasets.load_breast_cancer(return_X_y=True)
X = (X - X.mean(axis=0)) / (X.std(axis=0))

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state = 777, stratify = y)
```

```python
def cross_validation(X_train, y_train, n_split, ax):
    k_array = range(1, int((len(y_train)/n_split)*(n_split - 1)), 1)
    knn = KNeighborsClassifier()
    if n_split == len(y_train):
        kfold = LeaveOneOut()
    else:
        kfold = StratifiedKFold(n_split, shuffle=True, random_state=777)
        
    grid_cv = GridSearchCV(knn, param_grid={'n_neighbors':k_array}, cv=kfold,scoring='accuracy', n_jobs=-1)
    grid_cv.fit(X_train, y_train)

    ax.plot(1 - grid_cv.cv_results_['mean_test_score'], label = f'{n_split}-fold CV')

    min_cv_mean_err = 1 - grid_cv.best_score_
    k_opt = grid_cv.best_estimator_.n_neighbors
    print(f'{n_split}-fold CV k opt: {k_opt} and err: {min_cv_mean_err}')

    best_model = KNeighborsClassifier(n_neighbors=k_opt)
    best_model.fit(X_train, y_train)
    R_test = np.mean(y_test != best_model.predict(X_test))
    print(f'{n_split}-fold CV final test error: {R_test}')
    
    return grid_cv, best_model
```

```python
fig, ax = plt.subplots()

x1 = cross_validation(X_train, y_train, 5, ax)
x2 = cross_validation(X_train, y_train, 10, ax)
x3 = cross_validation(X_train, y_train, len(X_train), ax)

ax.set_xlabel('Neighbors, k')
ax.set_ylabel('CV Error')
ax.legend()
plt.show()
```

ошибки получились одинаковые, возможно из-за того что выборка не такая большая. во всех случаях k opt = 8. Но на более больших метод loo должен показывать себя наилучгим образом.

```python
k_opt = 8
best_model = KNeighborsClassifier(n_neighbors=k_opt)
best_model.fit(X_train, y_train)
y_test_predict = best_model.predict(X_test)

print(f'accuracy: {accuracy_score(y_test, y_test_predict)}')
print(f'f1: {f1_score(y_test, y_test_predict)}')
print(f'AUC: {roc_auc_score(y_test, best_model.predict_proba(X_test)[:,1])}')

fig, ax = plt.subplots()
RocCurveDisplay.from_estimator(best_model, X_test, y_test, ax=ax)
plt.title('ROC-кривая')
plt.show()
```

Задан вектор-признак **x** (независимая переменная) и вектор **y**, который представляет собой следующую зависимость от x: 

$$y = 3\cdot x-2$$

```python
x = [1, 4, 5, 3, 6, 8, 2, 11, 10, -1, 12, -4, 15, 16, -3, 0, 11, -2, 13, 7, -6]
y = [1, 10, 13, 7, 16, 22, 4, 31, 28, -5, 34, -14, 43, 46, -11, -2, 31, -8, 37, 19, -20]
```

Допустим, что мы не знаем, как связаны заданные **x** и **y** (забыли коэффициенты 3 и -2) и хотим построить линейную регрессионную модель вида $y = a\cdot x + b$. Необходимо изпользовать метод наименьших квадратов (решить систему линейных уравнений на слайде 7 лекции), чтобы определить неизвестные коэффициенты $a$ и $b$.

Совпали ли найденные коэффициенты с фактическими коэффициентами?

Вычислите остаточную сумму квадратов с найденными $a$ и $b$. Сделайте выводы.

```python
import numpy as np
import numpy.linalg as linalg
```

```python
def matrixA_b(x, y):
    sum_sqr_x = sum([i ** 2 for i in x])
    sum_x = sum(x)
    sum_dot_xy = sum([x[i]*y[i] for i in range(len(x))])
    N = len(x)
    sum_y = sum(y)
    A = [[sum_sqr_x, sum_x], [sum_x, N ]]
    B = [sum_dot_xy, sum_y]
    return A, B
```

```python
matrix_A, rigt_part = matrixA_b(x, y)
ab = linalg.lstsq(matrix_A, rigt_part, rcond=None)[0]
print(f'Коеффициент a {ab[0]}, b {ab[1]}')
```

Если округлить до целого, то коеффициенты полностью совпали

```python
RSS = sum([y[i] - (ab[0] * x[i] + ab[1]) for i in range(len(x))])
print(RSS)
```

Остаточная сумма получилась 0, следовательно мы отлично подобрали параметры.

```python
df = pd.read_csv('students_knowledge_status.txt', sep=';')

X = df.iloc[:, :-1]
X = (X - np.mean(X)) / np.std(X, ddof=1)

encoder = LabelEncoder()
encoder.fit(df.UNS)
y = encoder.transform(df.UNS)
```

```python
#1
from sklearn.cluster import KMeans

X_for_1 = X[['PEG', 'STR']]
km_model = KMeans(n_clusters=4, random_state=777)
km_model.fit(X_for_1)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 5))

ax1.scatter(X['PEG'], X['STR'], c = y, cmap='rainbow')
ax1.set(xlabel='PEG', ylabel='STR')
ax1.set_title('Известные классы учеников')

ax2.scatter(X['PEG'], X['STR'], c = km_model.labels_, cmap='rainbow')
ax2.set(xlabel='PEG', ylabel='STR')
ax2.set_title('Полученные кластеры')
```

Вывод: не соответсвует. скорее всего потому что этих 2х показателей не хватает для разбияния, которое хотим получить. так как классифицировал классификатор все хорошо разделяется.

```python
from scipy.cluster.hierarchy import linkage, dendrogram

methods = ['single', 'complete', 'average', 'weighted', 'centroid', 'median', 'ward']
fig, ax = plt.subplots(7, 1, figsize = (15, 60))

for i in range(len(methods)):
    ac_model = linkage(X, method=methods[i])    
    ax[i].set_title(f'Dendrogram for method {methods[i]}')
    dendrogram(ac_model, labels=df.UNS.values, leaf_rotation=90, leaf_font_size=10, ax=ax[i])
    
plt.show()
```

Вывод: наиболее удачным на мой взгляд является ward
