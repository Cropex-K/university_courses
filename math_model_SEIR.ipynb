```python
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
#уравнение стандартной модели 
#beta - частота контактов (скорость передачи)
#sigma - инкубационные период (как долго в состоянии E)
#gamma - период восстановления

def SEIR(t, y, beta, sigma, gamma):
    S, E, I, R = y
    dS_dt = - beta * I * S
    dE_dt = beta * I * S - sigma * E
    dI_dt = beta * E - gamma * I
    dR_dt = gamma * I
    return dS_dt, dE_dt, dI_dt, dR_dt
```

```python
b = 1
s = 1
g = 0.1

sol = solve_ivp(SEIR, [0, 60], [0.99, 0.1, 0, 0],rtol=1e-6, args=(b, s, g))

fig = plt.figure(); ax = fig.gca()
curves = ax.plot(sol.t, sol.y.T)
ax.legend(curves, ['S', 'E', 'I', 'R']);
```
